```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import math
from cmdstanpy import CmdStanModel
```

```python
snp = pd.read_json('../json_snp.json')
snp['d_mean'] = snp['d_mean']*100
snp['d_var'] = snp['d_var']*10000
print(snp)
```

```python
json_snp = snp.to_dict(orient = 'list')
json_snp['N_obs'] = len(snp)
json_snp['N_pop'] = 4
json_snp['ancestral_T'] = 800
```

```python
snp_model = CmdStanModel(stan_file="snp_alone.stan")
```

```python
fit = snp_model.sample(data = json_snp)
```
